```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../vehicles_us.csv")
```

```python
# Histograma de odometer
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="odometer", bins=50, kde=False)
plt.title("Histograma: odometer")
plt.xlabel("odometer")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=50, kde=False)
plt.title("Histograma: price")
plt.xlabel("price")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="odometer", y="price")
plt.title("Odometer vs Price")
plt.xlabel("Odometer")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="model_year", y="price")
plt.title("Precio vs Año del modelo")
plt.xlabel("Año del modelo")
plt.ylabel("Precio")
plt.show()
```
